```python
# ANÁLISIS EXPLORATORIO DE DATOS

# Se define la función para el análisis exploratorio de datos
def analisis_exploratorio(df):
    """
    Realiza el análisis exploratorio de datos para cada columna del DataFrame.

    Args:
        df (pd.DataFrame): El DataFrame con los datos.
    """

    resultados = {}  # Diccionario para almacenar los resultados del análisis

    for columna in df.columns:
        if pd.api.types.is_numeric_dtype(df[columna]):  # Solo para columnas numéricas
            media = df[columna].mean() # Se calcula la media
            maximo = df[columna].max() # Se calcula el máximo
            minimo = df[columna].min() # Se calcula el mínimo
            rango = maximo - minimo # Se calcula el rango
            desviacion_estandar = df[columna].std() # Se calcula la desviación estándar

            # Almacenar los resultados en el diccionario
            resultados[columna] = {
                "Media": media,  # Se calcula la media
                "Máximo": maximo, # Se calcula el máximo
                "Mínimo": minimo, # Se calcula el mínimo
                "Rango": rango, # Se calcula el rango
                "Desviación Estándar": desviacion_estandar # Se calcula la desviación estándar
            }

            # Imprimir los resultados
            print(f"Análisis de {columna}:") # Se imprime el nombre de la columna
            for estadistica, valor in resultados[columna].items(): # Se imprime el nombre de la estadística y su valor
                print(f"  {estadistica}: {valor}") # Se imprime el nombre de la estadística y su valor

            # Graficar el histograma
            plt.figure(figsize=(8, 6))
            sns.histplot(df[columna], kde=True)  
            plt.title(f"Histograma de {columna}")
            plt.xlabel("Valor")
            plt.ylabel("Frecuencia")
            plt.show()
    
    return pd.DataFrame(resultados).T  # Devuelve los resultados como DataFrame



#  Ejecutar el análisis exploratorio
resultados_df = analisis_exploratorio(df_training.copy())  # Usamos df.copy() para evitar modificar el original
print("\nResumen del análisis exploratorio:")
print(resultados_df.to_markdown(numalign="left", stralign="left"))
```

```python
# ANÁLISIS DE CORRELACIÓN E INTERACCIÓN ENTRE VARIABLES

# Se define la función para calcular la correlación y graficar
def calcular_correlacion_y_graficar(df, nombre_variable_dependiente='SalePrice', num_variables_a_mostrar=2):
    """
    Calcula la correlación entre cada variable independiente y la variable dependiente,
    genera scatterplots, y selecciona las N variables con mayor correlación.

    Args:
        df (pd.DataFrame): El DataFrame con los datos.
        nombre_variable_dependiente (str): El nombre de la columna que es la variable dependiente.
        num_variables_a_mostrar (int): El número de variables con mayor correlación a mostrar.
    """

    correlaciones = {} # Diccionario para almacenar las correlaciones
    
    # Calcular la correlación entre cada variable independiente y la variable dependiente
    for columna in df.columns:
        if columna != nombre_variable_dependiente and pd.api.types.is_numeric_dtype(df[columna]):
            correlacion = df[columna].corr(df[nombre_variable_dependiente])
            correlaciones[columna] = correlacion

            plt.figure(figsize=(8, 6))
            plt.scatter(df[columna], df[nombre_variable_dependiente])
            plt.title(f"Correlación entre {columna} y {nombre_variable_dependiente} (Correlación: {correlacion:.2f})")
            plt.xlabel(columna)
            plt.ylabel(nombre_variable_dependiente)
            plt.show()


    # Seleccionar las N variables con mayor correlación absoluta
    top_correlaciones = sorted(correlaciones.items(), key=lambda x: abs(x[1]), reverse=True)[:num_variables_a_mostrar]
    nombres_top_variables = [variable for variable, _ in top_correlaciones]

    print("\nTop Variables Correlacionadas con ", nombre_variable_dependiente, ":")
    for variable, correlacion in top_correlaciones:
        print(f"  {variable}: {correlacion:.2f}")
    
    return nombres_top_variables

#  Ejecutar el análisis de correlación y obtener las mejores variables
top_variables = calcular_correlacion_y_graficar(df_training.copy()) # Usamos df.copy() de nuevo
print("\nVariables seleccionadas para el modelo:", top_variables)
```

```python
# Función para entrenar el modelo manualmente
def entrenar_modelo_manual(x, y, epochs, imprimir_error_cada, learning_rate):
    """
    Función para entrenar el modelo de regresión lineal univariable.

    Argumentos:
        x : Es el vector con la variable independiente.
        y : Es el vector con la variable dependiente.
        epochs : Número de iteracciones de entrenamiento.
        imprimir_error_cada: Frecuencia para imprimir el error.
        learning_rate: Tasa de aprendizaje.

    Retornos:
        historial_modelos : Diccionario con los parámetros (beta0, beta1) de cada iterración.
        errores: Array con el error de cada iteracción.
    """

    # Creación de la matriz X con una columna de 1s
    X = np.column_stack((x, np.ones(len(x))))  # Añadir una columna de 1s
    n = len(y)  # Número de observaciones

    # Inicialización de los parámetros (beta0 y beta1)
    parametros = np.zeros(2)  # [beta1, beta0]
    historial_modelos = {}
    errores = np.zeros(epochs)

    # Iteracciones de entrenamiento
    for epoch in range(epochs):
        
        # Calcular la predicción de y_hat
        y_hat = np.dot(X, parametros)

        # Calcular el error (costo)
        error = (1 / (2 * n)) * np.sum((y - y_hat) ** 2)
        errores[epoch] = error

        # Calcular los gradientes
        gradiente = (1 / n) * np.dot(X.T, (y_hat - y))

        # Actualizar los parámetros
        parametros = parametros - learning_rate * gradiente

        # 3.5 Almacenar el modelo
        historial_modelos[epoch] = parametros.copy()  # Guardar una copia

        # Imprimir el error cada 'imprimir_error_cada' iteraciones
        if (epoch + 1) % imprimir_error_cada == 0:
            print(f"Epoch {epoch + 1}/{epochs}, Error: {error}")

    return historial_modelos, errores
```

```python
# Función para graficar los errores vs las iteracciones del entrenamiento manual
def graficar_error_vs_iteracion(errores):
    """
    Grafica el error en función del número de iteración.

    Args:
        errores (np.array): Array con los errores de cada iteración.
    """

    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(errores) + 1), errores)
    plt.title('Error vs. Iteración')
    plt.xlabel('Iteración')
    plt.ylabel('Error')
    plt.grid(True)
    plt.show()
```

```python
# Función para graficar la evolución del modelo de entrenamiento manual
def graficar_evolucion_modelo(df, historial_modelos, x, y, n):
    """
    Grafica la evolución del modelo entrenado a lo largo del tiempo (iteraciones).

    Args:
        df (pd.DataFrame):  DataFrame original (para los datos).
        historial_modelos (dict): Diccionario con los parámetros del modelo en cada iteración.
        x (str): Nombre de la columna usada para la variable independiente.
        y (str): Nombre de la columna usada para la variable dependiente.
        n (int): Frecuencia para graficar el modelo (cada n iteraciones).
    """

    plt.figure(figsize=(8, 6))
    plt.scatter(df[x], df[y], label='Datos', alpha=0.5)  #  Datos originales
    plt.title('Evolución del Modelo')
    plt.xlabel(x)
    plt.ylabel(y)

    num_iteraciones = len(historial_modelos)
    for i in range(0, num_iteraciones, n):
        beta1, beta0 = historial_modelos[i]
        y_pred = beta1 * df[x] + beta0
        plt.plot(df[x], y_pred, label=f'Iteración {i + 1}')

    plt.legend()
    plt.show()
```

```python
# Predicción con el modelo scikit-learn
from sklearn.linear_model import LinearRegression

def entrenar_modelo_sklearn(x, y):
    """
    Entrena un modelo de regresión lineal univariable usando scikit-learn.

    Args:
        x (np.array): Vector con la variable independiente (escalado).
        y (np.array): Vector con la variable dependiente (escalado).

    Returns:
        sklearn.linear_model.LinearRegression: El modelo entrenado.
    """

    modelo_sklearn = LinearRegression()  # Crear el modelo
    modelo_sklearn.fit(x.reshape(-1, 1), y)  # Entrenar el modelo
    return modelo_sklearn
```

```python

# Función para realizar el entrenamiento del modelo manualmente
def predecir_combinado(modelo_manual, modelo_sklearn, x):
    """
    Calcula las predicciones usando el modelo entrenado manualmente, el modelo de scikit-learn,
    y luego promedia las dos predicciones.

    Args:
        modelo_manual (tuple):  (beta0, beta1) del modelo entrenado manualmente.
        modelo_sklearn (sklearn.linear_model.LinearRegression): Modelo entrenado con scikit-learn.
        x (np.array): Vector de la variable independiente (escalado).

    Returns:
        tuple: (y_pred_manual, y_pred_sklearn, y_pred_combinado)
            - y_pred_manual (np.array): Predicciones del modelo manual.
            - y_pred_sklearn (np.array): Predicciones del modelo de scikit-learn.
            - y_pred_combinado (np.array): Predicciones combinadas (promedio).
    """

    beta0_manual, beta1_manual = modelo_manual
    y_pred_manual = beta1_manual * x + beta0_manual

    y_pred_sklearn = modelo_sklearn.predict(x.reshape(-1, 1))

    y_pred_combinado = (y_pred_manual + y_pred_sklearn) / 2

    return y_pred_manual, y_pred_sklearn, y_pred_combinado

```

```python
# Entrenamiento combinado para la variable OverallQual

# Seleccionar las variables que se utilizaran el entrenamiento
variable_independiente_nombre = 'OverallQual' 
variable_dependiente_nombre = 'SalePrice'
variable_independiente = df_training[variable_independiente_nombre].values
variable_dependiente = df_training[variable_dependiente_nombre].values


# Escalar los datos de la variable independiente
media_x = np.mean(variable_independiente)
std_x = np.std(variable_independiente)
variable_independiente_escalada = (variable_independiente - media_x) / std_x

# Escalar los datos de la variable dependiente
media_y = np.mean(variable_dependiente)
std_y = np.std(variable_dependiente)
variable_dependiente_escalada = (variable_dependiente - media_y) / std_y


#####################################################################################
# Entrenar los modelos
#####################################################################################


# Hiperparámetros de entrenamiento
epochs = 10000
imprimir_error_cada = 1000
learning_rate = 0.00001


# Entrenar el modelo manualmente
historial_modelos, errores = entrenar_modelo_manual(
variable_independiente_escalada,
variable_dependiente_escalada,
epochs,
imprimir_error_cada,
learning_rate
)
modelo_manual_final = historial_modelos[epochs - 1] # Últimos parámetros


# Entrenar el modelo con scikit-learn
modelo_sklearn = entrenar_modelo_sklearn(
variable_independiente_escalada,
variable_dependiente_escalada
)


#####################################################################################
# Hacer y mostrar predicciones
#####################################################################################


# Hacer predicciones combinadas
y_pred_manual_escalada, y_pred_sklearn_escalada, y_pred_combinado_escalada = predecir_combinado(
modelo_manual_final,
modelo_sklearn,
variable_independiente_escalada
)


# Desescalar las predicciones
y_pred_manual = y_pred_manual_escalada * std_y + media_y
y_pred_sklearn = y_pred_sklearn_escalada * std_y + media_y
y_pred_combinado = y_pred_combinado_escalada * std_y + media_y


# Imprimir las primeras predicciones
print("\nPrimeras 5 predicciones:")
print(pd.DataFrame({
'Original': df_training[variable_dependiente_nombre][:5].values,
'Manual': y_pred_manual[:5],
'Sklearn': y_pred_sklearn[:5],
'Combinado': y_pred_combinado[:5]
}))


# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(variable_independiente, variable_dependiente, label='Datos originales', alpha=0.5)
plt.plot(variable_independiente, y_pred_manual, color='red', label='Manual')
plt.plot(variable_independiente, y_pred_sklearn, color='green', label='Sklearn')
plt.plot(variable_independiente, y_pred_combinado, color='purple', label='Combinado')
plt.legend()
plt.title(f'Predicciones para {variable_independiente_nombre} vs. {variable_dependiente_nombre}')
plt.xlabel(variable_independiente_nombre)
plt.ylabel(variable_dependiente_nombre)
plt.show()


#####################################################################################
# Evaluación del modelo con los datos de prueba
#####################################################################################

# Preparar datos de prueba (¡Escalar con las medias y desviaciones del ENTRENAMIENTO!)
x_test = (df_test[variable_independiente_nombre].values - media_x) / std_x
y_test = df_test[variable_dependiente_nombre].values


y_test_pred_manual_escalada = modelo_manual_final[0] * x_test + modelo_manual_final[1]
y_test_pred_sklearn_escalada = modelo_sklearn.predict(x_test.reshape(-1, 1)).flatten()
y_test_pred_combinado_escalada = (y_test_pred_manual_escalada + y_test_pred_sklearn_escalada) / 2


y_test_pred_manual = y_test_pred_manual_escalada * std_y + media_y
y_test_pred_sklearn = y_test_pred_sklearn_escalada * std_y + media_y
y_test_pred_combinado = y_test_pred_combinado_escalada * std_y + media_y


def calcular_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))


rmse_manual = calcular_rmse(y_test, y_test_pred_manual)
rmse_sklearn = calcular_rmse(y_test, y_test_pred_sklearn)
rmse_combinado = calcular_rmse(y_test, y_test_pred_combinado)


print("\nRMSE en el conjunto de prueba:")
print(f" Manual: {rmse_manual:.2f}")
print(f" Sklearn: {rmse_sklearn:.2f}")
print(f" Combinado: {rmse_combinado:.2f}")


```
